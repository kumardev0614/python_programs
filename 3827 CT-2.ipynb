```python
import random as r
```

```python
def generate_vals():
    return {
            "d":r.choice(range(100,801)),
            "k":r.choice([4,5,6,7,8]),
            "l":r.choice([3,4,5,4,5])
           }
```

```python
generate_vals()
```

```python
final_list = []
count = 0
for i in range(1000):
    p = generate_vals()
    
    if (p['k'] != p['l']) and (p['d']%p['k'] == 0) and ((p['d']/p['k']) % p['l'] == 0):
        final_list.append(p)
        count = count + 1
        print('i = ', i, 'count =', count, "   {} % {}={} and     {} % {}={}".format(p['d'],p['k'],int(p['d']/p['k']),int(p['d']/p['k']),p['l'],int((p['d']/p['k'])/p['l'])))
        print("added", p)
        print("_______________________________________________")
        if count == 30:
            break     
```
